https://en.wikipedia.org/wiki/Multi-label_classification for multi-label classification problem

exact match loss:
Is the most strict metric, indicating the percentage of samples that have all their labels classified correctly.

hamming score loss:the fraction of the wrong labels to the total number of labels.This is a loss function, so the optimal value is zero.

```python
def hamming_loss(familyPre,familyTure,genusPre,genusTure,speciesPre,speciesTure):
    count=0;
    length_f=len(familyPre)
    length_g=len(genusPre)
    length_s=len(speciesPre)   
    for i in range(0,length_f):
        if(familyPre[i]==familyTure[i]):
            count=count+1
    for i in range(0,length_g):
        if(genusPre[i]==genusTure[i]):
            count=count+1
    for i in range(0,length_s):
        if(speciesPre[i]==speciesTure[i]):
            count=count+1
    total=length_f+length_g+length_s
    result=1-count/total
    return (result)
def exactMatch_loss(familyPre,familyTure,genusPre,genusTure,speciesPre,speciesTure):
    count=0;
    length=len(familyPre)
    for i in range(0,length):
        if(familyPre[i]==familyTure[i]):
            if(genusPre[i]==genusTure[i]):
                if(speciesPre[i]==speciesTure[i]):
                    count=count+1
    total=length
    result=1-count/total
    return (result)
```

## b_ii 

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.multiclass import OneVsRestClassifier #one vs rest classifier 

trainSetX=trainDf.drop(['Family','Genus','Species','RecordID'],axis=1)#get the data of train set
testSetX=testDf.drop(['Family','Genus','Species','RecordID'],axis=1)#get the data of test set

trainSetY_Family=pd.DataFrame(trainDf[["Family"]])
trainSetY_Genus=pd.DataFrame(trainDf[["Genus"]])
trainSetY_Species=pd.DataFrame(trainDf[["Species"]])

testSetY_Family=pd.DataFrame(testDf[["Family"]])
testSetY_Genus=pd.DataFrame(testDf[["Genus"]])
testSetY_Species=pd.DataFrame(testDf[["Species"]])

```

```python
gammanumber = np.array([value for value in range(1,20,2)])/7
est_gamma = 1/(2*gammanumber*gammanumber)
```

```python
parameters = {'estimator__C':[0.01,0.1,1,10,100],'estimator__gamma':est_gamma}
kernal = OneVsRestClassifier(SVC())#use one vs rest classifier
                                                        
clf_Family = GridSearchCV(kernal,parameters,cv = 10,n_jobs=-1)#cross validation is 10
clf_Genus = GridSearchCV(kernal,parameters,cv = 10,n_jobs=-1)
clf_Species = GridSearchCV(kernal,parameters,cv = 10,n_jobs=-1)

Family_model=clf_Family.fit(trainSetX,trainSetY_Family)#train label family
Genus_model=clf_Genus.fit(trainSetX,trainSetY_Genus)#train label genus
Species_model=clf_Species.fit(trainSetX,trainSetY_Species)#train label species
```

```python
predict_family=Family_model.predict(testSetX)
predict_genus=Genus_model.predict(testSetX)
predict_species=Species_model.predict(testSetX)

score_family=Family_model.score(testSetX,testSetY_Family)
score_genus=Genus_model.score(testSetX,testSetY_Genus)
score_species=Species_model.score(testSetX,testSetY_Species)

print('the correct rate of faimly label is '+ str(score_family))
print('the correct rate of genus label is '+ str(score_genus))
print('the correct rate of species label is '+ str(score_species))
```

```python
weight_Family=Family_model.best_params_
weight_Genus=Genus_model.best_params_
weight_Speices=Species_model.best_params_

sigma_family=math.sqrt((1/weight_Family['estimator__gamma']))
sigma_genus=math.sqrt((1/weight_Genus['estimator__gamma']))
sigma_speices=math.sqrt((1/weight_Speices['estimator__gamma']))

print("The weight of svm_Family is ",weight_Family,"sigma is ",sigma_family)
print("The weight of svm_Genus is ",weight_Genus,"sigma is ",sigma_genus)
print("The weight of svm_Speices is ",weight_Speices,"sigma is ",sigma_speices)
```

```python
predict_family=Family_model.predict(testSetX)
predict_genus=Genus_model.predict(testSetX)
predict_species=Species_model.predict(testSetX)

ture_family=testSetY_Family.values
ture_genus=testSetY_Genus.values
ture_species=testSetY_Species.values

hamming_loss_l1=hamming_loss(predict_family,ture_family,predict_genus,ture_genus,predict_species,ture_species)
exactMatch_loss_l1=exactMatch_loss(predict_family,ture_family,predict_genus,ture_genus,predict_species,ture_species)

print("The hamming loss of l1_penalized is ",hamming_loss_l1)
print("The exactMatch loss of l1_penalized is ",exactMatch_loss_l1)
```
